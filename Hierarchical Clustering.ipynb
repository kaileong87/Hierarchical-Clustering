```python
df_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.heatmap(data=scaled_df, cmap='magma')
```

```python
sns.clustermap(scaled_df, row_cluster=False)
```

```python
sns.clustermap(scaled_df, col_cluster=False)
```

Looking at the clustermap above, 

```python
from sklearn.cluster import AgglomerativeClustering
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.dendrogram.html

Let's assume every point is a cluster and we'll explore picking a distance threshold subsequently

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
labels = model.fit_predict(scaled_df)
```

```python
labels
```

Creating a linkage model

```python
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
plt.figure(figsize=(12,8), dpi=200)
dn = hierarchy.dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20,10), dpi=200)
dn = hierarchy.dendrogram(linkage_matrix, truncate_mode='lastp', p=6)
```

```python
plt.figure(figsize=(20,10), dpi=200)
dn = hierarchy.dendrogram(linkage_matrix, truncate_mode='level', p=6)
```
